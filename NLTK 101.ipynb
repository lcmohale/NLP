# NLTK Basics

Tokenization

Tokenization may be defined as the process of splitting the text into smaller parts called tokens, and is considered a crucial step in NLP.

```python
import nltk
import slate3k as slate
from nltk.tokenize import sent_tokenize
```

```python
pdfDocument = slate.PDF(open('stories-for-parents-children-and-grandchildren-volume-2.pdf','rb'))
```

```python
#The book is formatted in such a way that, a story is represented as a paragraph in a page
#I want the stories so...

stories = []

for i in range(len(pdfDocument)):
    stories.append(pdfDocument[i])
    
#Find the most frequent word in a story
def wordCount(story):
    import re 
    stopWords = nltk.corpus.stopwords.words('english')
    
    pat ='\w+'
    story = re.findall(pat,story)
    justStory = []
    
    for w in story:
        if w.lower() not in stopWords:
            justStory.append(w.lower())        
    
    freqDist = {}
    
    for word in justStory:
        if word not in freqDist:
            freqDist[word] = 1
        else:
            freqDist[word] += 1
            
    maximum = 0 
    key = ''
    for word in freqDist.keys():
        if freqDist[word] > maximum:
            maximum = freqDist[word]
            key = word
        
    return "The most frequent word in the text provided is {} it appears {} times".format(key, maximum)

for i in range(len(stories)):
    print("{} : {}".format(i, wordCount(stories[i])))

```
